# Popularity based recomender system

```python
# Merge ratings and books based on "ISBN"
ratings_with_book = ratings.merge(books, on='ISBN')
```

```python
ratings_with_book
```

```python
ratings_with_book['Book-Rating'].value_counts().head(10).plot(kind='bar')
```

```python
# We group by books title and count 
num_rating_df = ratings_with_book.groupby('Book-Title')['Book-Rating'].count().reset_index() # returns are new df
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
# We group by books title and count 
avg_rating_df = ratings_with_book.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
# Lets merge num_rating_df and avg_rating_df
popular_df = num_rating_df.merge(avg_rating_df, on= 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >=250].sort_values('avg_rating', ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Year-Of-Publication' ,'Image-URL-M', 'num_ratings', 'avg_rating']]
```

```python
# Export pickle file
import pickle

pickle.dump(popular_df, open('popular.pkl', 'wb'))
```

# Collaborative recomender system

```python
x = ratings_with_book.groupby('User-ID')['Book-Rating'].count() >200
ed_user = x[x].index # boolean indexing and then do .index
```

```python
# Filtering
filtered_rating = ratings_with_book[ratings_with_book['User-ID'].isin(ed_user)]
```

```python
y = filtered_rating.groupby('Book-Title')['Book-Rating'].count() >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
similarity_score
```

```python
def recommend(book_name):
    # index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:10]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('1984')
```

```python
pickle.dump(pt, open('pt.pkl', 'wb'))
pickle.dump(books, open('books.pkl', 'wb'))
pickle.dump(similarity_score, open('similarity_score.pkl', 'wb'))
```
